```python
b = np.array([12,2,455,1,8,8,765,4,32,3,4567,89,76,5432,4,5])
```

```python
np.argsort(b)
```

## Argmin (returns the smallest element)

```python
np.argmin(b)
```

## Argmax (returns the largest element)

```python
np.argmax(b)
```
