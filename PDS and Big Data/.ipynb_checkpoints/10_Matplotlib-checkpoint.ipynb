```python
import matplotlib.pyplot as plt
```

```python
plt.title("Statistics")
plt.ylabel("Y Label")
plt.xlabel("X Label")
plt.plot([1,2,-3],[-2,5,7])
plt.show()
```

```python
x1 = [10, 40, 60]
y2 = [12, 43, 76]
x2 = [12, 45, 86]
y2 = [54, 64, 98]

plt.plot(x1, y2, label='first')
plt.plot(x2, y2, label='second')
plt.legend()
#Need to use plt.legend() to use labels and it looks cool

plt.show()
```

```python
from matplotlib import style
style.use("ggplot")
plt.plot([1,2,3],[2,5,7], linewidth='10')
plt.show
```

```python
plt.bar([1,2,4],[2,5,7])
```

```python
plt.scatter([1,2,4,21,23,12,23,43],[2,5,7,12,27,43,42,53])
```

```python
slices = [7, 9, 1, 2]
activities = ["work", "sleep", "music", "games"]
cols = ['c', 'm', 'r', 'b']
plt.pie(slices,  #Mandatory(value for each pie)
       labels=activities,  #Names for each pie
       colors=cols,  #Colors for each pie
       startangle=90,  #To start dividing the first at 90deg angle
       shadow=True,  #To bring shadow effect
       explode=(0, 0, 0.1, 0), #will move out the 3rd piece by 0.1
       autopct='%1.1f%%') #For displaying %
plt.show
```
