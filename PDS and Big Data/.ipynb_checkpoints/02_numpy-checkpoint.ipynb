NUMPY IS FASTER THAN PYTHON AND ALSO TAKES LESS SPACE WHICH IS IMPORTANT FOR DATA ANALYSIS

Numpy is written in C

```python
import numpy as np
```

Numpy Array

```python
arr = np.arange(1000000)
```

```python
%time for _ in range(10): arr*=3
```

Python Array

```python
pythlist = list(range(1000000))
```

```python
%time for _ in range(10): [item*3 for item in pythlist]
```

converting python list to np array
