```python
arr = np.array([1, 2, 3, 4, 5])
print(arr+arr)
```

## Pointers

When an array or list is copied, it is not duplicated but instead a pointer is pointing to the original array, which is faster.

```python
arr2 = arr[1:3]
arr2[0] = 1952
```

```python
print(arr2)
print(arr)
```

If you dont want to set a pointer then you have to use the copy function

```python
arr = np.array([1, 2, 3, 4, 5])
arr2 = arr.copy()
arr2[0] = 1231
```

```python
print(arr2)
print(arr)
```
