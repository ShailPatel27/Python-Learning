## Matrix

```python
b = np.array([[10, 20],
              [30, 40], 
              [50, 60]])
np.dot(a, b)
```

```python
c = np.array([[12, 43, 103], 
             [87, 54, 9]])
# a.dot(c)  #will throw an error because the matrix needs to be transposed
```

```python
np.dot(a, c.transpose())
```

```python
np.cross(a, c)
```
